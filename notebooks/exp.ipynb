```python
import pymongo
import pandas as pd

df=pd.read_csv("data/spotify.csv")
df.head()
```

```python
spotify_dict=df.to_dict(orient="records")
```

```python
import os
DATABASE_NAME="spotify"
COLLECTION_NAME:str="spotifydata"
MONGODB_URL_KEY=os.getenv("MONGODB_URL")

```

```python
MONGODB_URL_KEY
```

```python
client=pymongo.MongoClient(MONGODB_URL_KEY)
database=client[DATABASE_NAME]
collention=database[COLLECTION_NAME]
```

```python
client.list_database_names()
```

```python
collention.insert_many(spotify_dict)
```

```python
loaded_df=pd.DataFrame(list(collention.find()))
```

```python
import time
import requests

url = "http://localhost:5000/predict"

payload = {
    "danceability": 0.808,
    "energy": 0.626,
    "key": 7,
    "loudness": -12.733,
    "mode": 1,
    "speechiness": 0.168,
    "acousticness": 0.00187,
    "instrumentalness": 0.159,
    "liveness": 0.376,
    "valence": 0.369,
    "tempo": 123.988,
    "duration_ms": 293027,
    "time_signature": 4,
    "chorus_hit": 35,
    "sections": 9
}

for i in range(10):
    start = time.time()
    response = requests.post(url, data=payload)  
    end = time.time()
    print(f"Response {i+1}: {response.status_code}, Time: {end-start:.3f}s")
```
